```python
from sympy import *
```

```python
x=Symbol('x')
fun = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
fun
```

**–ü–µ—Ä–∏–æ–¥**

```python
periodicity(fun, x)
```

**1.–û–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∫–æ—Ä–Ω–∏**


```python
list_solves =solve(fun,x)
list_solves 
```

```python
solveset(fun,x)
```

**4.–ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫**


```python
graf = plot(fun,(x,-pi,pi))
```

**–ü—Ä–æ–∏–∑–≤–æ–¥–Ω–∞—è**

```python
proizvodn = diff(fun)
```

```python
f=(-sin(ùë•)+cos(ùë•))*(sin(ùë•)/2+cos(ùë•)/2)*sin(ùë•)+(sin(ùë•)/2+cos(ùë•)/2)**2*cos(ùë•)
f
```

```python
graf =plot(diff(fun),(x,-pi,pi),line_color='red')
```

```python
list_diff = solve(diff(fun),x)
#list_diff.sort()  #—Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞
# —Ç–∞–∫ –∫–∞–∫ –ø–µ—Ä–∏–æ–¥–∏—á–Ω–æ—Å—Ç—å 2–ø–∏ - –¥–æ–±–∞–≤–ª—è–µ–º +1 –∫–æ—Ä–µ–Ω—å —á—Ç–æ –±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏–µ/—É–±—ã–≤–∞–µ–Ω–∏–µ –Ω–∞ –≤—Å–µ–º –ø–µ—Ä–∏–æ–¥–µ 
# –Ω–∞ —Å–µ–º–∏–Ω–∞—Ä–µ –æ–±—Å—É–∂–¥–∞–ª–∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –¥–≤—É—Ö –∫–æ—Ä–Ω–µ–π - –Ω–æ –æ–Ω –Ω—É–∂–µ–Ω –æ–¥–∏–Ω, –ø—Ä–∏ –¥–≤—É—Ö –±—É–¥–µ—Ç –ø–æ–≤—Ç–æ—Ä –ø–µ—Ä–≤–æ–π –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–π —á–∞—Å—Ç–∏
list_diff.append(list_diff[0]+2*pi) 
# —Ç—É—Ç –ø—Ä–æ—Å—Ç–æ –ø–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ü–∏—Ñ—Ä—ã –∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–∞ –≤—Å—è–∫–∏–π —Å–ª—É—á–∞–π 
list_diff_evalf=list_diff[:]
for i in range(len(list_diff)):
    list_diff_evalf[i] = list_diff[i].evalf()
list_diff_evalf.sort()

list_diff_evalf
```

**2.–ù–∞–π—Ç–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö —Ñ—É–Ω–∫—Ü–∏—è –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç** + **3.–ù–∞–π—Ç–∏ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö —Ñ—É–Ω–∫—Ü–∏—è —É–±—ã–≤–∞–µ—Ç**

```python

for i in range(len(list_diff_evalf)-1):
    y=proizvodn.subs(x,(list_diff_evalf[i]+list_diff_evalf[i+1])/2.0)
    
    if y>0:
        print(f'–≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ ({list_diff_evalf[i]},{list_diff_evalf[i+1]})')
    else:
        print(f'—É–±—ã–≤–∞–µ—Ç –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–∫–µ ({list_diff_evalf[i]},{list_diff_evalf[i+1]})')
    
```

**5.–í—ã—á–∏—Å–ª–∏—Ç—å –≤–µ—Ä—à–∏–Ω—É**

```python
print('–í–µ—Ä—à–∏–Ω—ã')
for i in range(len(list_diff)-1):#-1 –ø–æ—Ç–æ–º—É —á—Ç–æ –º—ã –ø—Ä–∏ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è –¥–æ–±–∞–≤–ª—è–ª–∏ –≤ –∫–æ–Ω–µ—Ü –∫–æ—Ä–µ–Ω—å
    y=fun.subs(x,list_diff[i])
    print(f'{i+1}.y = {y}  x = {list_diff[i]}')
   
```

```python
solve(fun>0,x)
```
