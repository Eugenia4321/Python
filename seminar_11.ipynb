```python
from sympy import *
```

```python
x=Symbol('x')
fun = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
fun
```

**Период**

```python
periodicity(fun, x)
```

**1.Определить корни**


```python
list_solves =solve(fun,x)
list_solves 
```

```python
solveset(fun,x)
```

**4.Построить график**


```python
graf = plot(fun,(x,-pi,pi))
```

**Производная**

```python
proizvodn = diff(fun)
```

```python
f=(-sin(𝑥)+cos(𝑥))*(sin(𝑥)/2+cos(𝑥)/2)*sin(𝑥)+(sin(𝑥)/2+cos(𝑥)/2)**2*cos(𝑥)
f
```

```python
graf =plot(diff(fun),(x,-pi,pi),line_color='red')
```

```python
list_diff = solve(diff(fun),x)
#list_diff.sort()  #сортировка
# так как периодичность 2пи - добавляем +1 корень что бы определить возрастание/убываение на всем периоде 
# на семинаре обсуждали добавление двух корней - но он нужен один, при двух будет повтор первой и последней части
list_diff.append(list_diff[0]+2*pi) 
# тут просто посмотрим на цифры и отсортируем на всякий случай 
list_diff_evalf=list_diff[:]
for i in range(len(list_diff)):
    list_diff_evalf[i] = list_diff[i].evalf()
list_diff_evalf.sort()

list_diff_evalf
```

**2.Найти интервалы, на которых функция возрастает** + **3.Найти интервалы, на которых функция убывает**

```python

for i in range(len(list_diff_evalf)-1):
    y=proizvodn.subs(x,(list_diff_evalf[i]+list_diff_evalf[i+1])/2.0)
    
    if y>0:
        print(f'возрастает на промежутке ({list_diff_evalf[i]},{list_diff_evalf[i+1]})')
    else:
        print(f'убывает на промежутке ({list_diff_evalf[i]},{list_diff_evalf[i+1]})')
    
```

**5.Вычислить вершину**

```python
print('Вершины')
for i in range(len(list_diff)-1):#-1 потому что мы при определении возрастания добавляли в конец корень
    y=fun.subs(x,list_diff[i])
    print(f'{i+1}.y = {y}  x = {list_diff[i]}')
   
```

**6.Определить промежутки, на котором f > 0**

```python
solve(fun>0,x)
```
